```python
import csv
import pandas as pd
```

```python
df = pd.read_csv(r'.\Cultivos.csv',sep=';',header=0,skiprows=1)\
  .dropna().drop(columns=['Cosechas-año/Distancia'])
df.rename(columns={'Tipo': 'tipo',\
  'Cultivo':'cultivo',
  'Min':'t_min',
  'Ideal':'t_opt',
  'Max':'t_max',
  'Cosecha 1 (días)':'germinacion_1',
  'Cosecha n (días)':'germinacion_n',
  'Distancia (cm)':'distancia',
  'Primera cosecha (mes)':'primera_cosecha',
  'N cosecha/año':'n_cosechas',
  'Distancia/Cosechas-año':'distanciaXcosechaAnual'},inplace=True)
df.head()
```

```python
with open(r'.\Data.csv','w+',encoding='utf-8') as file:
  txt = 'content:Cultivos\n'
  txt += df.to_csv(index=False,encoding='utf-8',escapechar='\n')
  txt += '\n'
  file.write(txt)
```
